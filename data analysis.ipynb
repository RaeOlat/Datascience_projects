```python
from pywaffle import Waffle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import squarify  


# Import dataset
Heart_disease = pd.read_csv('Heartdataset.csv')
print(Heart_disease) 

```

```python
# Plotting correlation matrix for dataset  
Heart_disease.corr()
```

```python
# Visualising dataset using heatmap
plt.figure(figsize = (16,5))
sns.heatmap(Heart_disease.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
plt.title("Heart disease dataset heatmap", fontsize =20)
```

```python
# Pairwise plot to visualise dataset
Heart_numerical_cat = Heart_disease.drop(columns =["sex","cp","fbs","restecg","exang","oldpeak","slope","ca","thal","target"])
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(Heart_numerical_cat, kind="reg")
plt.suptitle("Heart disease dataset pairwise plot", size=16, y=1.06);      

plt.show()
```

```python
# Strip plot for the numeric columns in the heart disease dataset
plt.figure(figsize = (16,5))
sns.stripplot(data=Heart_disease)

sns.stripplot(data=Heart_disease.drop(columns =["sex","cp","fbs","restecg","exang","oldpeak","slope","ca","thal","target"], axis=1))
plt.xlabel("Numerical categories for heart disease ")
plt.ylabel("Values")
plt.title("Heart disease dataset numerical categories")
```

```python
# Treemap for those diagnosed with heart disease and the rest that are healthy in the dataset
Heart_d = Heart_disease.groupby('target').size().reset_index(name='counts')
n_categories = Heart_d.shape[0]
colors = [plt.cm.inferno_r(i/float(n_categories)) for i in range(n_categories)]
print(Heart_d)
sizes=[138,165]

label=['Healthy','Diagnosed with heart disease']
colors=['green','blue']
squarify.plot(sizes, label=label, color=colors, alpha=0.5, pad=True)
plt.axis('off')
plt.show()
```
